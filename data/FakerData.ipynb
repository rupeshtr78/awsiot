```python
import csv
from random import randrange
from faker import Faker
fake = Faker()
```

```python
Faker.seed(0)
for _ in range(5):
  print(fake.first_name(),"|",
        fake.last_name(),"|",
#         fake.address(),"|",
        fake.phone_number(),"|",
        fake.license_plate(),"|",
        fake.random_int(min=60, max=110),
#         fake.location_on_land(),
        fake.longitude(),
        fake.latitude(),
        fake.street_address(),
        fake.zipcode(),
        fake.city(),
        fake.state()
      );
```

```python
def create_csv_file(record_count):
    with open('drives.csv', 'w', newline='') as csvfile:
        fieldnames = ['first_name', 'last_name', 'licensePlate', 'email', 'address',
                      'phoneNumber','longitude','latitude', 'city','state', 'speed']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for i in range(record_count):
            writer.writerow(
                {
                    'first_name': fake.first_name(),
                    'last_name': fake.last_name(),
                    'licensePlate': fake.license_plate(),
                    'email': fake.email(),
                    'address': fake.address(),
                    'phoneNumber': fake.phone_number(),
                    'longitude': fake.longitude(),
                    'latitude':fake.latitude(),
                    'city':fake.city(),
                    'state':fake.state(),
                    'speed': fake.random_int(min=60, max=110)
                  }
            )
```

```python
create_csv_file(1000)
```
